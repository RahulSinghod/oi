### 데이터 종류

#### 숫자형(Numerical Type) 데이터
* 연속성을 띄는 숫자로 이루어진 데이터
     * 예) Age, Fare 등

#### 범주형(Categorical Type) 데이터
* 연속적이지 않은 값(대부분의 경우 숫자를 제외한 나머지 값)을 갖는 데이터를 의미
    * 예) Name, Sex, Ticket, Cabin, Embarked 
* 어떤 경우, 숫자형 타입이라 할지라도 개념적으로 범주형으로 처리해야할 경우가 있음
    * 예) Pclass

### 함수를 이용하여 데이터 변환하기

```python
import math

def age_categorize(age):
    if math.isnan(age):
        return -1
    return math.floor(age / 10) * 10
```

```python
train_data.head()
```

```python
train_data['Age'].apply(age_categorize)
```

#### NaN 처리 방법
* 데이터에서 삭제
    * dropna 함수 
* 다른 값으로 치환
    * fillna 함수
        * 평균으로 대체하기
        * 생존자/사망자 별 평균으로 대체하기

```python
train_data.dropna().head()
```

```python
train_data.dropna(subset=['Age', 'Cabin']).head()
```

```python
train_data.dropna(axis=1).head()
```

```python
train_data['Age'].fillna(train_data['Age'].mean()).head()
```

```python
# 생존자 나이 평균
mean1 = train_data[train_data['Survived'] == 1]['Age'].mean()

# 사망자 나이 평균
mean0 = train_data[train_data['Survived'] == 0]['Age'].mean()

print(mean1, mean0)
```

```python
train_data[train_data['Survived'] == 1]['Age'].fillna(mean1).head()
```

```python
train_data[train_data['Survived'] == 0]['Age'].fillna(mean0).head()
```

```python
train_data.loc[train_data['Survived'] == 1, 'Age'] = train_data[train_data['Survived'] == 1]['Age'].fillna(mean1)
train_data.loc[train_data['Survived'] == 0, 'Age'] = train_data[train_data['Survived'] == 0]['Age'].fillna(mean0)
```
